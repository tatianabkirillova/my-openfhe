```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial import Polynomial
from numpy.polynomial import Chebyshev
from numpy.polynomial.chebyshev import chebfit

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-25, 25, 2**16)
y = sigmoid(x)

degree = 16

p = Polynomial.fit(x, y, degree)
c = Chebyshev.fit(x, y, degree)

p_coeffs = p.convert().coef
c_coeffs = c.convert().coef

p_coeffs
#c_coeffs
```

```python
c_coeffs
```

```python
y_poly = np.polyval(p_coeffs[::-1], x)
plt.plot(x, y, label='Sigmoid', lw=2)
plt.plot(x, y_poly, label=f'Polynomial Approximation (degree {degree})', linestyle='--', lw=2)
plt.legend()
plt.show()
```
